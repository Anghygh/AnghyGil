```python
# Problema 1: clasificador de temperatura segun el número
for i in range (-30,51):
    if i < 0:
      print(i, "Muy frío")
    elif i >= 0 and i < 24:
      print(i, "Templado")
    else:
      print(i, "Caluroso")
```

```python
# Problema 2: Números especiales del robot
for i in range(1,101):
  if i % 10 == 0:
    print(i, "Bip Bip")
  elif i % 5 == 0:
    print(i, "Bip")
```

```python
#Problema 3: detector de puntaje gamer
num1 = int(input("Ingrese el primer puntaje:"))
num2 = int(input("Ingrese el segundo puntaje:"))

suma = num1 + num2

if suma < 50:
  print("Nivel principiante")
elif suma >= 50 and suma < 80:
  print("Nivel intermedio")
else:
  print("Nivel pro")
```

```python
# Problema 4: Perfil Musical
edad = int(input("Ingrese su edad:"))
musica = input("¿Le gusta la música? (si/no):")
bailar = input("¿Te gusta bailar? (si/no):")

if edad >= 18 and musica == "si"  and bailar == "si":
  print("Eres una estrella de la fiesta")
elif musica == "si" and bailar == "no":
  print("Te gusta la música, pero no tanto bailar")
else:
  print("Tienes tu propio estilo")
```
